```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
df1 = pd.read_csv("student_data.csv")
```

```python
df1
```

```python
sns.catplot(x="gender", y="apps_phone", kind="violin", data=df1)
plt.show()
```

```python
sns.catplot(x='gender',y='apps_phone',data=df1)
plt.title('Amount of apps on phone by gender')
plt.ylabel('Amount of apps')
plt.xlabel('Gender')
plt.show()
```

```python
male = df1[df1['gender']=='male'] #this gets us a dataframe with only the paid apps
female = df1[df1['gender']=='female'] #ditto for the free apps

male_mean = male['apps_phone'].mean()
female_mean = female['apps_phone'].mean()
print(f'The mean amount of apps for male is {male_mean}')
print(f'The mean amount of apps for female apps is {female_mean}')
```

```python
ttest_ind(male['apps_phone'], female['apps_phone'], nan_policy='omit')
```

```python
# There is no significant relation between the amount of apps on phones of males or females.
```
