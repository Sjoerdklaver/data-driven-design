```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("class_data.csv")
```

```python
df
```

```python
sns.distplot(df['phones_owned'].dropna(), kde=False) #Selecting the rating column. I need to drop the NA's for the plot
plt.title('Phones Owned')
plt.xlabel('Amount')
plt.show()
```

```python
df1 = df["linkedin_con"]
df1 = df1.dropna()
```

```python
sns.distplot(df1)
plt.title('Linkedin Connections')
plt.ylabel('Proportion')
plt.show()
```

```python
df_apps_phone = df[df['apps_phone'] < 1000]
```

```python
sns.boxplot(df_apps_phone['apps_phone'])
plt.title('Number of Apps on Phone')
plt.xlabel('Amount of apps')
plt.show()
```

```python
df_paid = df[df['texts_yesterday'] < 900]
sns.violinplot(df_paid['texts_yesterday'])
sns.swarmplot(df_paid['texts_yesterday'], color='yellow')
plt.title('Amount of Texts Yesterday')
plt.ylabel('Frequency')
plt.xlabel('Texts')
plt.show()
```
