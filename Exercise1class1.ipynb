```python
import pandas as pd
```

```python
df1 = pd.read_csv("stepsdata.csv", sep = ';')
df2 = pd.read_csv("surveydata.csv")
df1.head(10)
```

```python
df2.head(10)
```

```python
df = pd.merge(df1, df2, on = "id")
df.head()
```

```python
df["waist"] = df["waist"].apply(pd.to_numeric, errors='coerce')
```

```python
df.head(10)
```

```python
df["waist"].value_counts()
```

```python
def above_110(x):
    if(x > 110.0): 
        return float('NaN')
    else: 
        return x
df["waist"] = df["waist"].apply(above_110)
```

```python
def under_65(x):
    if(x < 65.0): 
        return float('NaN')
    else: 
        return x
df["waist"] = df["waist"].apply(under_65)
```
